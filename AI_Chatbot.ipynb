```python
import os
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_groq import ChatGroq
```

```python
from google.colab import userdata

key = userdata.get("GROQ_API_KEY")

print("Key loaded:", key is not None)

os.environ["GROQ_API_KEY"] = key

```

```python
llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0.1
)
```

```python
parser= StrOutputParser()
```

```python
#hat_history list of messages
#message -> role and content

#role 1. system  2. user  3. assistance
```

```python
def chat():
    chat_history = [
        ("system", "You are a helpful chatbot. Be concise and accurate.")
    ]

    print("LangChain chatbot. Type 'exit' to quit.\n")

    while True:
        user_input = input("you: ").strip()

        if user_input.lower() == "exit":
            print("Goodbye ðŸ‘‹")
            break

        # Add user message
        chat_history.append(("human", user_input))

        # Build prompt from history
        prompt = ChatPromptTemplate.from_messages(chat_history)

        # Build chain
        chain = prompt | llm | parser

        # Get response
        response = chain.invoke({})

        # Show response
        print(f"bot: {response}\n")
        print("-" * 80)

        # Save assistant response
        chat_history.append(("assistant", response))
```

```python
chat()
```
